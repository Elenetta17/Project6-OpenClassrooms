L’entreprise "Place de marché” souhaite lancer une marketplace e-commerce.
Sur la place de marché, des vendeurs proposent des articles à des acheteurs en postant une photo et une description.
Pour l'instant, l'attribution de la catégorie d'un article est effectuée manuellement par les vendeurs, et est donc peu fiable. De plus, le volume des articles est pour l’instant très petit.

Pour rendre l’expérience utilisateur des vendeurs (faciliter la mise en ligne de nouveaux articles) et des acheteurs (faciliter la recherche de produits) la plus fluide possible, et dans l'optique d'un passage à l'échelle, il devient nécessaire d'automatiser cette tâche.

**Il faut donc réaliser une première étude de faisabilité d'un moteur de classification d'articles, basé sur une image et une description, pour l'automatisation de l'attribution de la catégorie de l'article.**

Je vais analyser le jeu de données en réalisant un prétraitement des descriptions des produits et des images et extraire des features pour effectuer ensuite un clustering. Je vais présenter les résultats du clustering sous la forme de graphiques en deux dimensions et calculer la similarité entre les catégories réelles et les clusters. 

Ces résultats confirmeront si les caractéristiques extraites permettent de regrouper des produits de même catégorie.


# 0. Imports and functions communes

```python
from PIL import Image
from sklearn import cluster
from sklearn import metrics
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.metrics import adjusted_rand_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder, MinMaxScaler


import gensim
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import time
```

```python
def clustering_scores(data, true_categories, N):
    """Cette fonction initialise N KMeans avec un n_clusters=7 et
    un random_state compris entre 0 et N, calcule le ARI entre
    vraies catégories et clusters et retourne le ARI moyen
    et le meilleur ARI avec le random_state correspondant"""
    ari_scores = []
    for i in range(N):
        kmeans = KMeans(n_clusters=7, random_state=i)
        clusters = kmeans.fit_predict(data)
        ari_scores.append(adjusted_rand_score(true_categories, clusters))
    ari_scores = np.array(ari_scores)
    print('ARI moyen: ' + str(ari_scores.mean()))
    print('ARI max: ' + str(ari_scores.max()))
    print('random_state de l\'ARI max: ' + str(ari_scores.argmax()))
```

## 2.0 Imports et fonction communes pour l'extraction et clustering des features texte

```python
from collections import Counter
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
def tokenize(text):
    """Cette fonction prend en input une string, la tokenise
    à l'aide de la fonction word_tokenize et retourne la liste
    de tokens composés exclusivement que des lettres"""
    return [token.lower() for token in word_tokenize(text) if token.isalpha()]
```

```python
def lemmatize_tokens(tokens):
    """Cette fonction prend en entrée une liste de tokens et retourne la liste
    des tokens lemmatisés"""
    lemmatizer = WordNetLemmatizer()
    return [lemmatizer.lemmatize(token) for token in tokens]
```

## 2.1 Pré-traitements

### 2.1.1 Tokenisation

La fonction `tokenize` transforme les noms des produits en liste de tokens. Dans cette fonction, en utilisant la méthode `isalpha`, on exclut de la liste tous le tokens qui contiennent de la ponctuation et/ou des chiffres. En effet, comme l'on peut voir dans les exemples de noms de produit ci-dessous, les tokens qui contiennent des chiffres pourraient se référer au modèle spécifique du produit, à sa taille ou encore au nombre de composantes du produit (p.e. Pack of 2). Il ne s'agit pas forcement d'informations pertinentes et ces tokens pourraient introduire du bruit lors du clustering.

```python
# initialisation de la variable tokens
data['tokens'] = data['product_name'].apply(tokenize)
```

## 2.2 Extractions de features et premiers tests de clustering

### 2.2.1 Bag of Words - comptage simple

Le principe de Bag of Words est le suivant: un document particulier, dans mon cas le nom d'un produit, sera représenté par l'histogramme des occurrences des mots le composant : pour un document donné, chaque mot se voit affecté le nombre de fois qu'il apparaît dans le document. Un document sera donc représenté par un vecteur de la même taille que le dictionnaire, dont la composante i indique le nombre d'occurrences du i-ème mot du dictionnaire dans le document. 

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# instanciation du CountVectorizer
cvect = CountVectorizer(stop_words=stopwords_all, min_df=4)
```

```python
# vectorisation des mots
bow = cvect.fit_transform(data['lemmatized_tokens'].apply(lambda tokens : ' '.join(tokens)))
```

```python
# initialisation dataframe des vecteurs de mots
bow = pd.DataFrame(bow.todense(), columns=cvect.get_feature_names_out(), index=data.index)
```

```python
# échantillon vecteurs de mots
bow.sample(5)
```

Maintenant que j'ai une matrice numérique représentant les noms des produits, je peux l'utiliser pour faire un clustering. La fonction `clustering_score` effectue plusieurs KMeans, avec un nombre de clusters = 7 (comme le nombre de catégories) et différents initialisations des centroïdes et retourne l'ARI moyen et l'ARI maximal (avec le random_state correspondant)

```python
clustering_scores(bow, data['category'], 100)
```

### 2.2.3 Word2vec

Word2vec est un groupe de modèles utilisé pour le plongement lexical (word embedding). Ce sont des réseaux de neurones artificiels à deux couches entraînés pour reconstruire le contexte linguistique des mots. Je vais entrainer un modèle Word2vec pour ensuite extraire la representation vectorielle des mots et l'utiliser dans le clustering.

```python
# paramètres entrainement modèle word2vec
w2v_size = 300
w2v_window = 5
w2v_min_count = 1
w2v_epochs = 100
```

```python
# initialisation et entraînement du modèle Word2Vec
print("Build & train Word2Vec model ...")
w2v_model = gensim.models.Word2Vec(min_count=w2v_min_count,
                                   window=w2v_window,
                                   vector_size=w2v_size,
                                   workers=1)
w2v_model.build_vocab(data['tokens'])
w2v_model.train(data['tokens'],
                total_examples=w2v_model.corpus_count,
                epochs=w2v_epochs)
model_vectors = w2v_model.wv
w2v_words = model_vectors.index_to_key
print("Vocabulary size: %i" % len(w2v_words))
print("Word2Vec trained")
```

On peut voir si les embeddings sont raisonnables en examinant, par exemple, les mots les plus similaires à `laptop`.

```python
# mots les plus similaires à laptop
model_vectors.most_similar(positive=['laptop'])
```

En effet, on trouve des marques d'ordinateur et des accessoires pour ordinateurs.

Le modèle peut retourner les vecteurs associés à chaque mot, mais pas à une séquence de mots (comme ça peut l'être le nom d'un produit).
Ainsi, pour chaque nom de produit, je vais récupérer les vecteurs de mots qui le composent et faire la moyenne de ces vecteurs à l'aide de la fonction `vectorize_sentence_w2v`.

```python
def vectorize_sentence_w2v(tokenized_sentence):
    """Cette fonction prend en input une liste de tokens
    et retourne la moyenne des vecteurs associés à chaque token"""
    if tokenized_sentence == []:
        return np.zeros(300)
    return model_vectors[tokenized_sentence].mean(axis=0)
```

```python
# calcul des vecteur et initialisation du dataframe des vecteurs
vectorize_description = data['tokens'].apply(vectorize_sentence_w2v)
data_word2vec = pd.DataFrame(np.vstack(vectorize_description),
                             index=data.index)
```

```python
# échantillon données
data_word2vec.sample(5)
```

On a bien, pour chaque produit, le vecteur correpondant (qui fait une taille de 300).

```python
# scores ARI
clustering_scores(data_word2vec, data['category'], 100)
```

L'utilisation de words embeddings a permis une nette amélioration de l'ARI.

### 2.2.4 Bert

Bert est une autre approche de type plongement de mots. L’encodeur du modèle de BERT lit toute la séquence de mots en une fois. Cette caractéristique permet au modèle d’apprendre le contexte d’un mot en fonction de tout son environnement (gauche et droite du mot). Je vais utiliser un modèle BERT pre-entrainé pour extraire les embeddings.

```python
from transformers import BertTokenizer, TFBertModel
```

```python
# chargement du modèle et tokeniseur Bert
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = TFBertModel.from_pretrained("bert-base-uncased")
```

Je vais encoder la variable `product_name` à l'aide de la fonction `vectorize_sentence_bert`

```python
def vectorize_sentence_bert(text):
    """Cette fonction prend en input une string, la tokenise
    et retourne la string encodée"""
    encoded_input = tokenizer(text, return_tensors='np')
    output = model(encoded_input)
    return output
```

```python
# initialisation de la variable des embeddings
data['bert'] = data['product_name'].apply(vectorize_sentence_bert)
```

```python
# initialisation du dataframe des embeddings
data_bert = pd.DataFrame(np.vstack(data['bert'].apply(lambda x : x[1])), index = data.index)
```

```python
# échantillon données
data_bert.sample(5)
```

On a bien, pour chaque produit, le vecteur correpondant (qui fait une taille de 768).

```python
# scores ARI
clustering_scores(data_bert, data['category'], 100)
```

Le modèle BERT ne donne pas de bons résultats. Probablement, il nécessite de fine-tuning.

### 2.2.5 USE - Universal Sentence Encoder

USE utilise les mécanismes d’attention pour calculer des représentations vectorielles sensibles au contexte des mots. Je vais utiliser un modèle USE pre-entrainé pour extraire les embeddings.

```python
import tensorflow_hub as hub
```

```python
# chargement du modèle
embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")
```

Je vais encoder la variable `product_name` à l'aide de la fonction `vectorize_sentence_use`

```python
def vectorize_sentence_use(text):
    """Cette fonction prend en input une string,
    et retourne la string encodée"""
    return embed([text])
```

```python
# calcul des embeddings et initialisation du dataframe des embeddings
data_use = data['product_name'].apply(vectorize_sentence_use)
data_use = pd.DataFrame(np.vstack(data_use), index=data.index)
```

```python
# échantillon données
data_use.sample(5)
```

## 3.0 Imports et fonctions communes pour l'extraction et clustering des features images

```python
from keras.models import Model
from keras.utils import img_to_array
from keras.utils import load_img

import cv2
```

```python
"""fonctions pour la détection des keypoints (SIFT ou ORB) et la 
création de visual word (histogramme), adaptées du notebook associé au webinaire"""
def find_keypoints_and_descriptors(path, names, method, filter_type=None):
    # identification of key points and associated descriptors
    keypoints = []
    temps1 = time.time()

    if method == 'SIFT':
        detector = cv2.xfeatures2d.SIFT_create(500)
    elif method == 'ORB':
        detector = cv2.ORB_create(500)

    for image_num in range(len(names)) :
        if image_num % 100 == 0 : print(image_num)
        image = cv2.imread(path+names[image_num],0) # convert in gray
        if filter_type == 'Gaussian':
            image = cv2.GaussianBlur(image, (3, 3), 0)
        if filter_type == 'Median':
            image = cv2.medianBlur(image, 3)
        res = cv2.equalizeHist(image)   # equalize image histogram
        kp, des = detector.detectAndCompute(res, None)
        keypoints.append(des)

    keypoints_by_img = np.asarray(keypoints)
    keypoints_all = np.concatenate(keypoints_by_img, axis=0)

    print()
    print("Nombre de descripteurs : ", keypoints_all.shape)

    duration1 = time.time() - temps1
    print("temps de traitement SIFT descriptor : ", "%15.2f" % duration1, "secondes")
    return keypoints_by_img, keypoints_all

def build_histogram(kmeans, des, image_num):
    # construction du mot visuel
    res = kmeans.predict(des)
    hist = np.zeros(len(kmeans.cluster_centers_))
    nb_des=len(des)
    if nb_des == 0 : print("problème histogramme image  : ", image_num)
    for i in res:
        hist[i] += 1.0/nb_des
    return hist

def feature_creation(keypoints_all, keypoints_by_img, names):
    # Determination number of clusters
    temps1=time.time()

    k = int(round(np.sqrt(len(keypoints_all)), 0))
    print("Nombre de clusters estimés : ", k)
    print("Création de",k, "clusters de descripteurs ...")

    # Clustering
    kmeans = cluster.MiniBatchKMeans(n_clusters=k, init_size=3*k)
    kmeans.fit(keypoints_all)

    duration1 = time.time() - temps1
    print("temps de traitement kmeans : ", "%15.2f" % duration1, "secondes")
    # Creation of histograms (features)
    temps1=time.time()

    # Creation of a matrix of histograms
    hist_vectors=[]

    for i, image_desc in enumerate(keypoints_by_img) :
        if i%100 == 0 : print(i)  
        hist = build_histogram(kmeans, image_desc, i) #calculates the histogram
        hist_vectors.append(hist) #histogram is the feature vector

    im_features = np.asarray(hist_vectors)

    duration2 = time.time() - temps1
    print("temps de création histogrammes : ", "%15.2f" % duration2, "secondes")
    return pd.DataFrame(im_features, index = names)
```

```python
def image_vectorization_CNN(images_names, model, model_type):
    """Cette fonction prend en entrée une liste de noms d'images,
    un modèle pré-entrainé et le type du modèle.
    Pour chaque image, elle la charge, réalise les pré-traitements adaptés
    et extrait les features.
    La fonction retourne le dataframe des features de toutes les images"""
    result = []
    for filename in images_names:
        image = load_img('Data/Images/' + filename,
                         target_size=(224, 224))
        image = img_to_array(image)
        image = image.reshape((1,
                               image.shape[0],
                               image.shape[1],
                               image.shape[2]))
        if model_type == 'ResNet':
            image = ResNet_preprocess(image)
        if model_type == 'VGG':
            image = VGGpreprocess(image)
        result.append(model.predict(image))
    return pd.DataFrame(np.concatenate(result, axis=0),
                        index=images_names)
```
